```
import keras
```

```
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array
```

```
imggen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```
imggen
```

```
from keras.preprocessing.image import load_img
```

```
k = load_img('/content/drive/My Drive/images.jpg')
```

```
k
```

```
x = img_to_array(k)
```

```
x.shape
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(x)
```

```
x = x.reshape((1,) + x.shape)
```

```
x.shape
```

```
i =0
for batch in imggen.flow(x,batch_size = 1,save_to_dir = '/content/drive/My Drive/vishnu',save_prefix = 'cat',save_format = 'jpeg'):
  i+=1
  if i>20:
    break
```
