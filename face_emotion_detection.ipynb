```
import warnings
warnings.filterwarnings('ignore')
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
import face_recognition
import keras
from keras.models import load_model
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/surprise.jpg')
```

```
image
```

```
plt.imshow(image)
```

```
image = cv2.resize(image,(48,48))
```

```
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```

image = np.reshape(image,[1,image.shape[0],image.shape[1],1])
```

```
image
```

```
model = load_model('/content/drive/My Drive/model_v6_23.hdf5')
```

```
import warnings
warnings.filterwarnings('ignore')
```

```
image.shape
```

```
predicted_class = np.argmax(model.predict(image))
```

```
emotion_dictonary= {'Angry': 0, 'Sad': 5, 'Neutral': 4, 'Disgust': 1, 'Surprise': 6, 'Fear': 2, 'Happy': 3}
label_map = dict((v,k) for k,v in emotion_dictonary.items())
```

```
predicted = label_map[predicted_class]
```
